```python
%%time
p_data = [pd.merge(pd.read_csv(filepath + sense_files[i], parse_dates=['dev_dt', 'dt']), pd.read_csv(filepath + status_files[i], parse_dates=['dev_dt', 'dt']), on = ['dev_id', 'dev_dt'], how = 'outer', sort=True) for i in range(5)]
```

```python
def basic_statistics(name, data, col_name, is_number = True):
    if is_number:
        elements = [
            name,
            data[col_name].describe().name,
            '%d' % data[col_name].describe()['count'],
            data[col_name].isnull().sum(),
            data[col_name].value_counts(dropna=False).count(),
            data[col_name].dtype.name,
            '%.05f' % data[col_name].describe()['mean'],
            '%.05f' % data[col_name].describe()['std'],
            '%.05f' % data[col_name].describe()['min'],
            '%.05f' % data[col_name].describe()['25%'],
            '%.05f' % data[col_name].describe()['50%'],
            '%.05f' % data[col_name].describe()['75%'],
            '%.05f' % data[col_name].describe()['max'],
            '%.05f' % np.divide(data[col_name].describe()['std'], data[col_name].describe()['mean'])
        ]
    else:
        elements = [
            name,
            data[col_name].describe().name,
            '%d' % data[col_name].describe()['count'],
            data[col_name].isnull().sum(),
            data[col_name].value_counts(dropna=False).count(),
            data[col_name].dtype.name,
            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan
        ]    
    return elements
```

```python
%%time
stats = [basic_statistics('%s' % (p_data[number]['dev_id'].unique()[0]), p_data[number], name, np.issubdtype(p_data[number][name].dtype, np.number)) for number in range(len(p_data)) for name in p_data[number]]
stats_result = pd.DataFrame(stats, columns=['name','column', 'counts','missing','unique','type','mean','std','min','q1','median','q3','max','cv'])
```
